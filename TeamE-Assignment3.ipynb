```python
import pandas as pd
from matplotlib import pyplot as plt   #import key modules 
import numpy as np
```

### Load the data set 

```python
from sklearn.datasets import load_diabetes
diabetes_dataset = load_diabetes()  
```

```python
data = pd.DataFrame(diabetes_dataset['data'])
target = pd.DataFrame(diabetes_dataset['target'])
target = target.rename(columns={0:'Target'})
data.describe() # Describe the data set 
data['target'] = target
data
```

```python
pd.plotting.scatter_matrix(data, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

linear_r = LinearRegression()
features = [2,8,9]  # Only keeping features 2,8,9
X_train = data.iloc[25:,features]   
y_train = target.iloc[25:,0]
X_test = data.iloc[0:25,features]
y_test = target.iloc[0:25,0]
lm_r = linear_r.fit(X_train, y_train)
diabetes_predict1 = lm_r.predict(X_test)
print('Coefficients:', lm_r.coef_)
print('Intercept:', lm_r.intercept_)
print('Coefficient Of Determination: %.2f' % r2_score(y_test, diabetes_predict1))
```

```python
# Ridge Regression 
from sklearn.linear_model import Ridge

X_train = data.iloc[25:,features]
y_train = target.iloc[25:,0]
X_test = data.iloc[0:25,features]
y_test = target.iloc[0:25,0]

# By varying the alpha parameter we see that l
ridge = Ridge(alpha = .001)
ridge_fit = ridge.fit(X_train, y_train)
diabetes_predict2 = ridge_fit.predict(X_test)
print('Coefficients:', ridge_fit.coef_)
print('Intercept:', ridge_fit.intercept_)
print('Coefficient Of Determination: %.2f' % r2_score(y_test, diabetes_predict2))
```

```python
#Lasso Regression 
from sklearn.linear_model import Lasso
import numpy as np

lasso = Lasso(alpha = .001 , max_iter = 100000)
lasso_fit = lasso.fit(X_train, y_train)
diabetes_predict3 = lasso_fit.predict(X_test)
print('Coefficients:', lasso_fit.coef_)
print('Intercept:', lasso_fit.intercept_)
print('Coefficient Of Determination: %.2f' % r2_score(y_test, diabetes_predict3))
print('Number of features used:', np.sum(lasso_fit.coef_ != 0))
```

```python
plt.plot(linear_r.coef_, 's', label="Linear Regresssion")
plt.plot(ridge_fit.coef_, '^', label="Ridge alpha=0.001")
plt.plot(lasso_fit.coef_, 'v', label="Lasso alpha=0.001")

plt.legend(ncol=2, loc=(0, 1.05))
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.axhline(0, color = 'black', linewidth = 0.75)
plt.show()
```

```python
import pandas as pd
import io
import requests

names = []
for i in range(1,35):
    names.append('Pulse'+str(i))
names.append('Class')

url="https://archive.ics.uci.edu/ml/machine-learning-databases/ionosphere/ionosphere.data"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')), names = names)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss

Coefficients = []
Probablities = []
Intercept = []
Log_of_Probabilities = []
Train_Score = []
Test_Score = []
Log_Loss = []
Run = []

x =  c.loc[:, 'Pulse1':'Pulse34']
y = c.loc[:,'Class'].to_frame()

X_train = x.iloc[25:,:]   
y_train = y.iloc[25:,:]
X_test = x.iloc[0:25,:]
y_test = y.iloc[0:25,0]

c_value = [0.1,1,10,60,100]
penality = ['l2','l1']

for i in c_value:
    for j in penality:
        print("C = " + str(i))
        print("Penality = " + j)
        print("")
        logreg = LogisticRegression(C = i, penalty = j)
        logreg.fit(X_train,y_train.values.ravel())
        print("Coefficients:", logreg.coef_)
        print("Intercept:", logreg.intercept_[0])
        print("Probabilities: " ,logreg.predict_proba(X_test))
        print("Log of Probabilites: ", logreg.predict_log_proba(X_test))
        print("Train Score: ", logreg.score(X_train, y_train))
        print("Test Score: ", logreg.score(X_test, y_test))
        test_predictions = logreg.predict(X_test)
        print("Log-loss: ", log_loss(test_predictions, logreg.predict_proba(X_test)))
        
        Run.append('C = '+str(i)+ ' , Penality = ' + j) 
        Coefficients.append(logreg.coef_)
        Intercept.append(logreg.intercept_[0])
        Train_Score.append(logreg.score(X_train, y_train))
        Test_Score.append(logreg.score(X_test, y_test))
        Log_Loss.append(log_loss(test_predictions, logreg.predict_proba(X_test)))
        print('--------------------------------------------------------------------------------')
```

```python
table = {'Run': Run, 'Intercepts': Intercept, 'Train Score': Train_Score, 'Test Score': Test_Score, 'Log Loss': Log_Loss}
df = pd.DataFrame(table)

def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: green' if v else '' for v in is_max]

def highlight_min(s):
    is_min = s == s.min()
    return ['background-color: red' if v else '' for v in is_min]

values = df.style.apply(highlight_min).apply(highlight_max)
values
```
