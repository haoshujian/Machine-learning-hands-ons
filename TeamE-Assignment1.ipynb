### Split data into train test split 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 0, train_size = 142, test_size = 8)

# Visualize the data 
pd.plotting.scatter_matrix(X_train, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
plt.show()
```

#### Inputing the given dataset from the homework 

```python
X_test_hw = [[5.8, 2.7, 5.2, 2.1],  [6.3, 2.7, 4.4, 1.4],  [5.0, 3.0, 5.3, 2.7],  [6.5, 3, 3.4, 2.1],  [4.9, 2.9, 1.9, 1.7],  [5.8, 2.7, 3.9, 2.4],  [5.1, 3.1, 1.7, 1.2],  [4.8, 2.8, 1.8, 1.3]]
X_test = pd.DataFrame(X_test_hw, columns = iris_dataset['feature_names'])
```

```python
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier 
# number of neighbors
neighbors = [1,3,5,9,15]   #put the neighbors into a list 
# add the distance types into a list 
distance = ["manhattan", "euclidean", "minkowski"]
weight = "uniform" 

# create a function that takes the neighbors, distance, and weight 
# uses a loop to assess the K-NN model 
def run_knn(neighbors, distance, weight, X_test): 
    for j in distance:
        print("The type of distance used is: "+j+"\n")
        test_accuracy = []
        for i in neighbors:
            knn = KNeighborsClassifier(n_neighbors = i, metric = j, weights = weight)
            knn.fit(X_train, y_train.values.ravel())
            prediction = knn.predict(X_test)
            print("Model Paramters: \n {}".format(knn) + "\n")
            print("Test set predictions for "+str(i)+" neighbors: \n{}".format(prediction)  + "\n")
            print("Predicted target name: \n{}".format(iris_dataset['target_names'][prediction]) + "\n")
            score = knn.score(X_test, y_test)
            print("Test set accuracy for "+str(i)+" neighbors: {:.2f}".format(score))
            print("-------------------------------------------------------------")
            test_accuracy.append(score)
        
        plt.plot(neighbors, test_accuracy, label="test accuracy")
        plt.ylabel("Accuracy")
        plt.xlabel("neighbors")
        plt.legend() 
        plt.show()
        print("================================================================")
    
run_knn(neighbors, distance, weight, X_test)
```

```python
# number of neighbors

neighbors = [1,3,5,9,15]
distance = ["minkowski"]
weight = "distance"

run_knn(neighbors, distance, weight, X_test_hw)
```

#### Description of Results: 
In these examples we took the KNN model form scikit learn and passed it different paramters for neighbors, distance and weight. However in doing so we did not see that significant of a change in accuracy when comparing it to the actual results (y-Test). The following are the reasons we are seeing poor accuracy scores:
* Only 8 rows of X test data. Usually when you split training and test data you should do atleast a 75% / 25% split. That way you have many records to test against
* The test data provided in this homework was selected at random. Since this data was selected at random and not from actual measurements in the field, there is a small chance in returning high scores 
* The wieghts, neighbors, distance measurements did not play any role in the end. There is too much uncertainty with the data to create a bias from any of the paramters.
